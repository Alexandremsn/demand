apos vizualizar como os dados estão dispostos Vamos criar uma função para cruzar os dados das duas tabelas e me devolver a informação de todas as lojas afetadas com aquele feriado em particular.

```python
listnational = list(storesdf['store_nbr'])
dictstate = dict(zip(storesdf.store_nbr, storesdf.state))
dictcity = dict(zip(storesdf.store_nbr, storesdf.city))  
def stores_affected(locale, locale_name):
  if locale == 'National':
    return listnational
  elif locale == 'Regional':
    statelist = []
    for i in listnational:
      if  dictstate[i] == locale_name:
        statelist.append(i)
    return statelist     
  elif locale ==   'Local':
    citylist = []
    for i in listnational:
      if  dictcity[i] == locale_name:
        citylist.append(i)
    return citylist
  else:
    print('error')
```

```python
holidaysdf['stores_affected'] = holidaysdf.apply(lambda x: stores_affected(x['locale'],x['locale_name']), axis=1)
```

```python
holidaysdf= holidaysdf[holidaysdf['type']!='Work Day']
```

removemos os dias úteis e criamos uma função para clasificar os feirados

```python
def holiday_classifier(typehd, description):
  if typehd == 'Bridge':
    return 'is_holiday'
  elif typehd == 'Holiday':
    return 'is_holiday'
  elif typehd == 'Transfer':
    return 'is_holiday'
  elif typehd == 'Additional':
    if description.find('-')>=0:
      return 'is_holiday_eve'
    else:
      return 'is_additional'
  elif typehd == 'Event':
    if description.find('Mundial')>=0:
      return 'is_world_cup'
    elif description.find('Terremoto')>=0:
      return 'is_natural_disaster'
    elif description.find('Black')>=0:
      return 'is_black_friday'
    elif description.find('Cyber')>=0:
      return 'is_cyber_monday'
    else:
      return 'is_holiday'
```

```python
holidaysdf['classification'] = holidaysdf.apply(lambda x: holiday_classifier(x.type, x.description), axis=1)
```

```python
holidaysdf[holidaysdf['classification']=='is_holiday_eve'].head(50)
def major_holiday(classification, description):
  if classification == 'is_holiday_eve':
    if description.find('Navidad')>=0:
      return 'Navidad'
    elif description.find('Madre')>=0:
      return 'Madre'
    else:
      return 'None'
holidaysdf['major_holiday'] = holidaysdf.apply(lambda x: major_holiday(x.classification,x.description),axis=1)
```
