```python
import pandas as pd
from sklearn.impute import KNNImputer
oil = pd.read_csv('oil.csv')
test = pd.read_csv('test_reduce.csv')
imputer = KNNImputer(n_neighbors=5)
oil.dcoilwtico = imputer.fit_transform(oil[['dcoilwtico']])
from sklearn.neighbors import KNeighborsRegressor
print(oil.shape)
oil.date = pd.to_datetime(oil.date, format='%Y-%m-%d')
print(oil.info())
```

vamos limpar os dados do dataset do preço do petróleo, vamos imputar os valores faltantes com o KNNImputer, vamos converter a data de string para datetime

```python
testsize = int(oil.shape[0]*0.2)
```

```python
df_train = oil[:-testsize]
df_test = oil[-testsize:]
X_train = df_train[['date']]
X_test = df_test[['date']]
y_train = df_train['dcoilwtico']
y_test = df_test['dcoilwtico']
```

vamos separar nossos dados em teste e treino, vamos realizar vários teste variado o principal hiperparâmetro até obter um erro menor possível.
no dataset de teste não temos o valor pois se trata de uma data futura, assim vamos usar um modelo preditivo simples para prever este valor.

```python
modelbasic = KNeighborsRegressor(n_neighbors=29)
modelbasic.fit(X_train, y_train)
y_pred = modelbasic.predict(X_test)
from sklearn.metrics import mean_squared_error
import numpy as np
score = np.sqrt(mean_squared_error(y_test,y_pred))
print(score)
```

obtivemos um bom erro, após testar várias conposições de hiperparâmetros chegamos na acima demonstrada, 

```python
test.date = pd.to_datetime(test.date, format='%Y-%m-%d')
X_train2 = oil[['date']]
y_train2 = oil.dcoilwtico
X_test2 = test[['date']]
modelbasic.fit(X_train2,y_train2)
test['oil_price'] = modelbasic.predict(X_test2)
```

agora vamos prever e gerar a coluna com valor do óleo no dataset de teste. e depois exportar para uso na próxima etapa do projeto.

```python
test.to_csv('test_oil.csv', index=False)
```
